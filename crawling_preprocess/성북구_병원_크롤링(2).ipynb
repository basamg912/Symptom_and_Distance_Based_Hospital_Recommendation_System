```python
# 받은 진료 텍스트 추출
def extract_treat(main_treatment):
    try:
        treat = driver.find_elements(By.XPATH, "//h1[contains(@class, 'styles_servicedTreatment__6L_kF')]")
        # "받은 진료 : "를 제거하고 나머지 텍스트만 추출
        for t in treat:
            treatment_text = t.text.replace("받은 진료 : ", "").strip()
            if treatment_text:
                main_treatment.append(treatment_text)
            else:
                main_treatment.append("X")
            
    except:
        main_treatment.append("X")
```

```python
# 의사 이름 추출
def extract_doc(main_doc):
    try:
        doc = driver.find_elements(By.XPATH, "//p[contains(@class, 'styles_doctor__jkVo4')]")
        # "받은 진료 : "를 제거하고 나머지 텍스트만 추출
        for d in doc:
            docname = d.text.replace("의사 : ", "").strip()
            main_doc.append(docname)
    except:
        main_doc.append("X")
```

```python
# 리뷰 추출 함수 gpt 수정
def extract_review(main_review):
    #time.sleep(2)
    
    # 리뷰 개수 추출
    try:
        review_c = driver.find_element(By.XPATH, "//p[contains(@class, 'styles_amount__xW0Ui')]")
        review_count = int(review_c.text)
    except:
        review_count = 0
    
    # "더보기" 버튼을 모두 클릭
    try:
        more_buttons = driver.find_elements(By.XPATH, "//button[contains(@class,'styles_seeMore__oWZDz')]")
        for btn in more_buttons:
            try:
                driver.execute_script("arguments[0].click();", btn)
                time.sleep(0.1)  # 클릭 사이에 약간의 지연을 줌
            except Exception as e:
                print(f"더보기 클릭 실패: {e}")
    except:
        print("더보기 버튼이 없습니다.")
    
    # 리뷰 내용 추출
    try:
        rev = driver.find_elements(By.XPATH, "//pre[contains(@class, 'styles_content__29gm9 styles_showAll__lxO56')]")
        if len(rev) == 0:
            main_review.append("리뷰 없음")
        else:
            # "받은 진료 : "를 제거하고 나머지 텍스트만 추출
            for r in rev:
                revcontent = r.text.strip()
                main_review.append(revcontent)
    except Exception as e:
        print(f"리뷰 추출 실패: {e}")
        main_review.append("리뷰 없음")
```

```python
# 자세한 설명/ 치료후 결과 / 직원의 친절/ 청결함 순서 점수
def extract_score(score1,score2,score3,score4,user_rate):
    explain = []
    after_treat = []
    employee = []
    clean = []
    try:
        pre_container = WebDriverWait(driver, 2).until(
            EC.presence_of_all_elements_located((By.XPATH, "//pre[contains(@class, 'styles_content__29gm9')]"))
        )
        for review in pre_container:
            star1= review.find_elements(By.XPATH,".//div[contains(@class,'styles_container__z87VT')]")
            
            star_score = 0
            total_stars = 5
            score_list = []
            for i in star1:
                
                full_stars=0
                half_stars=0
                j=i.find_elements(By.XPATH,".//*[local-name()='svg']")
                for k in j:
                    paths = k.find_elements(By.XPATH,".//*[local-name()='path']")
                    if len(paths) >= 2:
                        # 첫 번째 path: 채워지지 않은 부분 (반쪽 별일 가능성 있음)
                        first_fill = paths[0].get_attribute("fill")
                        second_fill = paths[1].get_attribute("fill")
                        # 반쪽 별 조건: 첫 번째 fill이 #fff이고, 두 번째 fill이 current 또는 Amber30
                        if first_fill == "#fff" and second_fill in ["current", "var(--Amber30)"]:
                            half_stars += 0.5
                        elif second_fill in ["current", "var(--Amber30)"]:
                            full_stars += 1
                    else:
                        # 단일 path인 경우
                        fill_value = paths[0].get_attribute("fill")
                        if fill_value in ["current", "var(--Amber30)"]:
                            full_stars += 1
                    total_score = full_stars+half_stars
                score_list.append(total_score)
            if len(score_list) == 4:
                score1.append(score_list[0])
                score2.append(score_list[1])
                score3.append(score_list[2])
                score4.append(score_list[3])
    except:
        score1.append("x")
        score2.append("x")
        score3.append("x")
        score4.append("x")

    try:
        total_score = driver.find_elements(By.XPATH,"//div[contains(@class,'styles_score__aUlv9')]")
        for score in total_score:
            user_score=score.text
            user_rate.append(user_score)
    except:
        user_rate.append("x")
```

```python
# 전체 지출 비용 리스트 초기화
def extract_cost(main_cost, main_treat, main_ins):
    total_costs = []
    treatment_prices = []
    ins = []
    
    # 모든 `pre` 요소를 찾는다
    try:
        pre_containers = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, "//pre[contains(@class, 'styles_content__29gm9')]"))
        )
        
        for review in pre_containers:
            try:
                # `pre`의 다음 형제 `li` 내의 지출 비용 추출
                total_cost_element = review.find_element(By.XPATH, "./following-sibling::li[contains(@class, 'styles_container__w5D8a')]/p[contains(@class, 'styles_totalCost__FVozM')]/span")
                total_cost = total_cost_element.text
                main_cost.append(total_cost)
        
                
            except:
                # 지출 비용이 없는 경우
                main_cost.append(None)
        
            temp = []
            try:
                # `pre`의 다음 형제 `li` 내의 지출 비용 추출
                treatment_element = review.find_elements(By.XPATH, "./following-sibling::li[contains(@class, 'styles_container__w5D8a')]//p[contains(@class, 'styles_price__ebelK')]")
                for i in treatment_element:
                    temp.append(i.text)
                main_treat.append(temp)
            except:
                # 지출 비용이 없는 경우
                main_treat.append("x")
            try:
                insure = review.find_element(By.XPATH,"./following-sibling::li[contains(@class,'styles_container__w5D8a')]//p[contains(@class,'styles_treatmentSummary__JDrZ3')]")
                if ("보험" in insure.text):
                    main_ins.append("보험 적용")
            except:
                main_ins.append("x")
    except:
        main_cost.append("x")
        main_treat.append("x")
        main_ins.append("x")
```
