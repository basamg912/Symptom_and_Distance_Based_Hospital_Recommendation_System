```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
```

```python
def get_hospital_info(url, dept_name):
    driver.get(url)
    time.sleep(2)  # í•„ìš” ì‹œ WebDriverWaitìœ¼ë¡œ êµì²´
    doctors = []
    doctor_scores = [] 
    res = bool
    try:
        name = driver.find_element(By.CSS_SELECTOR, "h1.styles_hospitalName__NTHfd").text.strip()
    except:
        name = ""
    
    try:
        addr = driver.find_element(By.XPATH, "//address[contains(text(), 'ì„œìš¸')]").text.strip()
    except:
        addr = ""
    
    try:
        score = driver.find_element(By.CLASS_NAME, "styles_score__aUlv9").text.strip()
    except:
        score = ""
    
    try:
        review = driver.find_element(By.XPATH, "//button[contains(text(), 'ë¦¬ë·°')]").text.strip().replace("ë¦¬ë·° ", "")
    except:
        review = ""
    try:
        docs = driver.find_elements(By.XPATH, "//div[contains(@class,'styles_doctorInfo__')]//p")
        for doc in docs:
            doctors.append(doc.text)
    except:
        doctors = ""

    try:
        doc_scores = driver.find_elements(By.XPATH, "//div[contains(@class,'styles_doctorInfo__')]//div[contains(@class,'styles_score__')]")
        for doc_score in doc_scores:
            doctor_scores.append(doc_score.text)
    except:
        doc_scores = ""
        
    try:
        reserve = driver.find_elements(By.XPATH,"//section[contains(@class, 'styles_container__F0Twl')]//p[contains(@class,'styles_buttonText__klDAH')]")
        for i in reserve:
            if "ì˜ˆì•½í•˜ê¸°" in i.text:
                res = True
                break
            else:
                res = False
    except:
        res = False

    return {
        "ì§„ë£Œê³¼": dept_name,
        "ë³‘ì›ëª…": name,
        "ì£¼ì†Œ": addr,
        "í‰ì ": score,
        "ì˜ì‚¬": doctors,
        "ì˜¨ë¼ì¸ ì˜ˆì•½ ì—¬ë¶€" : res,
        "ì˜ì‚¬ í‰ì ": doctor_scores,
        "ë¦¬ë·°ìˆ˜": review,
        "ë³‘ì›ë§í¬": url
    }
```

```python
#options = Options()
#options.add_argument('--headless')  # ğŸ‘ˆ ë¸Œë¼ìš°ì € ì°½ ì•ˆ ë„ì›€
#options.add_argument('--disable-gpu')  # (Windows ì•ˆì •ì„± ìœ„í•´)
#options.add_argument('--window-size=1920x1080')  # í™”ë©´ í¬ê¸° ì„¤ì •
#driver = webdriver.Chrome(options=options)

driver = webdriver.Chrome()

driver.get("https://www.modoodoc.com/")
time.sleep(3)

try:
    more_btn = WebDriverWait(driver,2).until(
        EC.element_to_be_clickable((By.XPATH,"//span[text()='ë”ë³´ê¸°']/ancestor::a"))
    )
    more_btn.click()
    print("ë”ë³´ê¸° ë²„íŠ¼ í´ë¦­")
    time.sleep(1)
except:
    print("ì‰£")

dept_names = []
depts = driver.find_elements(By.CSS_SELECTOR, "span.styles_text__hOjOw")
for dept in depts:
    dept_name = dept.text.strip()
    dept_names.append(dept_name)
dept_names = [name for name in dept_names if name != "ì‹œë ¥êµì •ìˆ  ê²€ì§„"]
```

```python
def go_to_main():
    main_url = "https://www.modoodoc.com/"
    driver.get(main_url)
    try:
        more_btn = WebDriverWait(driver,2).until(
            EC.element_to_be_clickable((By.XPATH,"//span[text()='ë”ë³´ê¸°']/ancestor::a"))
        )
        more_btn.click()
        print("ë”ë³´ê¸° ë²„íŠ¼ í´ë¦­")
        time.sleep(1)
    except:
        print("ì‰£")
    time.sleep(3)
```

```python
no_hospital = []
def hospital_links(departments,name):
    departments = driver.find_elements(By.CSS_SELECTOR, "span.styles_text__hOjOw")
    for dept in departments:
        dept_name = dept.text.strip()
        if dept_name == name:
            print(f"{name} í´ë¦­")
            parent = dept.find_element(By.XPATH, "./ancestor::a")
            parent.click()
            time.sleep(2)
            WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//span[contains(text(),'ì „ì²´')]"))
            ).click()
    
            # ì„±ë¶êµ¬ í´ë¦­
            try:
                WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.XPATH, "//span[text()='ì„±ë¶êµ¬']/ancestor::button"))
                ).click()
            except:
                print("ì„±ë¶êµ¬ ë³‘ì› x")
                no_hospital.append(dept_name)
    
            try:
                WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.XPATH, "//p[text()='ë‹¤ì‹œ ë³´ì§€ ì•Šê¸°']/ancestor::button"))
                ).click()
            except:
                print("ë‹¤ì‹œ ë³´ì§€ ì•Šê¸° ì•ˆ ëœ¸")
            try:    
                driver.find_element(By.XPATH, "//button[contains(., 'ë°©ë¬¸')]").click()
                time.sleep(2)
                WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.XPATH, "//span[text()='ë§Œì¡±ë„ìˆœ']/ancestor::li"))
                ).click()
            except:
                print("ë°©ë¬¸ ìˆœ ì•„ë‹˜")
            time.sleep(2)
            
            while True:
                try:
                    btn = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.XPATH, "//p[contains(text(),'ë”ë³´ê¸°')]/ancestor::button"))
                    )
                    driver.execute_script("arguments[0].click();",btn)
                    time.sleep(2)
                except:
                    print("ìŠ¤í¬ë¡¤ í•  ê±° ì—†ìŒ")
                    break
        
            cards = driver.find_elements(By.TAG_NAME, 'a')
            hospital_links = set()
            for card in cards:
                href = card.get_attribute("href")
                if href and "/hospital/" in href:
                    hospital_links.add(href)
            print(f"{dept_name} ë§í¬ {len(hospital_links)}ê°œ ìˆ˜ì§‘")
            return list(hospital_links)
```

```python
results = []
departments = driver.find_elements(By.CSS_SELECTOR, "span.styles_text__hOjOw")
for name in dept_names:
    links = hospital_links(departments,name)
    for link in links:
        info = get_hospital_info(link, name)
        results.append(info)
    go_to_main()
```

```python
df = pd.DataFrame(results)
df
```
