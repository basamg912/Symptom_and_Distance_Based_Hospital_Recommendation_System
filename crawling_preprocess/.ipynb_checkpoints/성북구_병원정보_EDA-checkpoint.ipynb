```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

df = pd.read_excel('최종_성북구_병원.xlsx')
df
```

```python
df_count = df.groupby('진료과').size().reset_index(name='병원 수').sort_values(by='병원 수')
df_count = df_count.sort_values(by='병원 수')
```

```python
df_count
```

```python
plt.rcParams['font.family'] = 'AppleGothic'
plt.title("진료과별 병원 갯수")
plt.bar(df_count['진료과'], df_count['병원 수'])
plt.xticks(rotation=90, fontsize=10)
```

```python
# 소아과를 해당 진료과별로 다시 분류
df_count_modi = df_count.copy()

dept = df['진료과'].unique()
for i in df[df['진료과']=='소아과']['병원명']:
    for j in dept:
        if j in i:
            df_count_modi.loc[df_count_modi['진료과']==j, '병원 수'] +=1

df_count_modi = df_count_modi.drop(6)
df_count_modi = df_count_modi.sort_values(by='병원 수')
```

```python

plt.rcParams['font.family'] = 'AppleGothic'
plt.title("진료과별 병원 갯수")
plt.bar(df_count_modi['진료과'], df_count_modi['병원 수'])
plt.xticks(rotation=90, fontsize=10)
```

```python
import numpy as np

bins = np.arange(0,df['평점'].max() + 0.5, 0.5)
df['평점구간'] = pd.cut(df['평점'],bins, right=False)
grade_counts = df['평점구간'].value_counts().sort_index()
grade_counts
```

```python
grade_counts.plot(kind = 'bar', figsize=(12,6), color='skyblue')
plt.xlabel('평점구간',fontsize=12)
plt.ylabel('갯수',fontsize=12)
plt.title('평점구간 별 병원수',fontsize=14)
plt.xticks(rotation=39)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
