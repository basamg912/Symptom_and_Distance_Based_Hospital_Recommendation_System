```python
import ast

def parse_won_string(won_str):
    return int(won_str.replace('원', '').replace(',', '').strip())

def sum_costs(list_str):
    try:
        items = ast.literal_eval(list_str)
        if not items:
            return 0
        total = sum(parse_won_string(i) for i in items if i.strip() != '-' and i.strip() != '‘-’')
        return int(total)
    except:
        return 0

# 사용 예시
def total_costs(list_str):
    try:
        cleaned = str(list_str).replace('원', '').replace(',', '').strip()
        if not cleaned.isdigit():
            return 0
        return int(cleaned)
    except:
        return 0

print(total_costs("6,000원"))

print(sum_costs("['-', '-', '-', '-' ]"))         # 출력: 0
print(sum_costs("['8,400 원', '10,000원']"))      # 출력: 18400
print(sum_costs("[]")) 
```
